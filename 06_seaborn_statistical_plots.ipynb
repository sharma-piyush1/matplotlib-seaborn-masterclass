# Seaborn Statistical Plots

This notebook covers:
- Distribution plots
- Category vs Numeric comparisons
- Spread vs central tendency
- Which plot to use when comparing groups

Goal: Understand not just how to draw the plots, but how to interpret them.


```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
```

Key Idea:
Box plots answer how spread out the data is â€” not just averages.

3) Violin Plot (Shape of Distribution in Groups)

```python
sns.violinplot(data=tips, x="day", y="tip")
plt.title("Distribution of Tip Amounts by Day")
plt.show()
```

Violin plots = box plot + density info
Use when the shape matters.

4) Strip + Swarm (Sample-Level Detail)

Never forget: Averages hide reality.

```python
sns.swarmplot(data=tips, x="day", y="tip", color="black", size=3)
plt.title("Individual Tips by Day")
plt.show()
```

5) Bar Plot with Statistical Estimation

Barplots in seaborn are NOT raw values â€” they use mean by default.

```python
sns.barplot(data=tips, x="day", y="tip", estimator=sum)
plt.title("Total Tips by Day (Sum)")
plt.show()
```

Want median instead of mean?

```python
import numpy as np
sns.barplot(data=tips, x="day", y="tip", estimator=np.median)
plt.title("Median Tip by Day")
plt.show()
```

When to Use What (This is the actual knowledge)
| Goal                              | Best Plot   | Why                       |
| --------------------------------- | ----------- | ------------------------- |
| Compare distribution shapes       | Violin      | Shows density & variation |
| Compare group medians/variability | Box         | Clean summary + outliers  |
| Show individual data points       | Strip/Swarm | Shows noise & clusters    |
| Compare totals                    | Bar (sum)   | Aggregation               |
| Compare averages                  | Bar (mean)  | Central tendency          |
