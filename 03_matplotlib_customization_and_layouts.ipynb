# Layouts, Subplots & Professional Chart Composition

This notebook focuses on:
- Multi-plot layouts
- Shared axis scaling
- Tight spacing and aspect ratios
- Titles, annotations, legends, and readability

Goal: Create visuals that look intentional and clear.


1) Subplots with Meaningful Arrangement

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

axes[0].plot(x, y_sin, color='blue')
axes[0].set_title("Sine Wave")

axes[1].plot(x, y_cos, color='red')
axes[1].set_title("Cosine Wave")

plt.tight_layout()
plt.show()
```

Key Concept:

Shared axes makes comparisons easier.

2) Grid of Plots

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title("Plot 1")

axes[0, 1].plot(x, np.cos(x))
axes[0, 1].set_title("Plot 2")

axes[1, 0].plot(x, np.tan(x))
axes[1, 0].set_title("Plot 3")

axes[1, 1].plot(x, np.exp(x / 3))
axes[1, 1].set_title("Plot 4")

plt.tight_layout()
plt.show()
```

Skill:

You now control multiple axes with clarity.

3) Super Title for Whole Figure

```python
fig.suptitle("Trigonometric Function Overview", fontsize=18, y=1.02)
```

This goes after fig, axes = ... and before plt.tight_layout().

4) Legend Positioning (Professional Use)

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, y_sin, label="Sine")
ax.plot(x, y_cos, label="Cosine")

ax.legend(loc="upper right", frameon=False)
```

Always remove legend borders  clean.

5) Annotation (Storytelling)

This is where visualizations start showing insight, not just data.

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(x, y_sin)
ax.set_title("Key Insight Annotation Example")

peak_x = np.pi / 2
peak_y = np.sin(peak_x)

ax.annotate(
    "Local Max",
    xy=(peak_x, peak_y),
    xytext=(peak_x+1, peak_y+0.5),
    arrowprops=dict(arrowstyle="->", color="black")
)

plt.show()
```

Real Analysts Annotate

People who dont annotate are just plotting for decoration.
