# Seaborn Styling & Theming (Professional Visual Identity)

This notebook focuses on:
- Choosing and applying visual themes
- Using palettes correctly
- Controlling scale, contrast, and emphasis
- Building a consistent visual identity across the entire repo


1) Global Style Setup

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", context="notebook")
```

Available styles:
| Style       | Feel                               |
| ----------- | ---------------------------------- |
| `whitegrid` | clean, analytical (use most often) |
| `darkgrid`  | data-heavy dashboards              |
| `white`     | minimal reports                    |
| `dark`      | rare, highlight mode               |


2) Palette Selection (The Single Most Abused Part)

You don’t choose colors you like.
You choose colors based on contrast and interpretability.

```python
sns.color_palette("deep")
sns.color_palette("muted")
sns.color_palette("pastel")
sns.color_palette("bright")
sns.color_palette("dark")
sns.color_palette("colorblind")
```

Use cases:

muted → reports

deep → dashboards

colorblind → universal accessibility

bright/pastel → avoid for professional

Default palette: deep → safe and clean.

3) Apply Palette Globally

```python
sns.set_palette("deep")
```

4) Contrast Control (Context Scaling)

```python
sns.set_context("paper")   # smallest
sns.set_context("notebook")  # default
sns.set_context("talk")     # bigger text
sns.set_context("poster")   # huge text
```

Use:

notebook for repo work

talk when preparing interview presentations

5) Example of Professional Styling in Practice

```python
tips = sns.load_dataset("tips")

sns.set_theme(style="whitegrid", palette="deep", context="notebook")

sns.barplot(data=tips, x="day", y="total_bill", hue="sex")

plt.title("Average Total Bill by Day and Gender", pad=12)
plt.xlabel("Day of Week")
plt.ylabel("Average Total Bill ($)")
plt.legend(title="Gender", frameon=False)
plt.show()
```

Notice:

Minimal contrast

Titles with spacing

Legend without border

Readable without loud colors

This looks corporate, not amateur.

6) Turning Off Borders & Spines (Advanced Clean Look)

```python
sns.despine()
```

Used to remove unnecessary lines that clutter your figure.
