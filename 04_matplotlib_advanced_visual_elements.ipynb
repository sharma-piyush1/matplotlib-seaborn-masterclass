# Advanced Visual Elements (Professional Visualization Layer)

In this notebook, we learn:
- Secondary y-axes (dual-scale charts)
- Highlighting key ranges (shaded regions)
- Custom grid control
- Controlled color and transparency
- Confidence interval visualization

Goal: Communicate insights clearly and intentionally.


1) Twin Axes (Dual Y-Scale Plot)

Used constantly in:

Finance (price vs volume)

HR analytics (headcount vs hiring trend)

Sales (revenue vs conversion)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y_main = np.exp(x / 3)
y_secondary = np.sin(x)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(x, y_main, color='blue', label="Growth Curve")
ax1.set_ylabel("Growth", color='blue')

ax2 = ax1.twinx()
ax2.plot(x, y_secondary, color='red', label="Cyclic Pattern")
ax2.set_ylabel("Cycle", color='red')

ax1.set_title("Dual Y-Axis Visualization Example")

plt.show()
```

Key Understanding:

You now know how to plot two different units cleanly on one chart.

2) Shading Regions to Highlight Important Ranges

Used for:

Highlighting business thresholds

Showing before/after effects

Marking significant zones in time series

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x, np.sin(x), linewidth=2)

ax.axvspan(2, 4, color='red', alpha=0.15, label='Event Window')
ax.set_title("Shaded Range to Highlight Important Interval")
ax.set_xlabel("Time")
ax.set_ylabel("Signal")

plt.show()
```

3) Confidence Intervals (Without Seaborn)

This makes your visualizations look statistical, not decorative.

```python
y = np.sin(x)
noise = np.random.normal(0, 0.2, size=len(x))
y_noisy = y + noise

std = np.std(noise)

fig, ax = plt.subplots(figsize=(10,5))

ax.plot(x, y_noisy, color='black', alpha=0.7, label="Observed")
ax.plot(x, y, color='blue', linewidth=2, label="True Signal")

ax.fill_between(x, y - std, y + std, color='blue', alpha=0.2, label="1 Std Dev")

ax.legend()
ax.set_title("Confidence Interval Visualization")
plt.show()
```

4) Professional Color Choice Principles

### Color Principles for Professional Plots
- Use **desaturated colors** (no neon garbage)
- Use 1 bold accent color per figure
- Use gray for secondary elements
- Avoid more than 5 distinct colors in one view
