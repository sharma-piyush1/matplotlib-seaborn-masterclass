```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
x = np.linspace(0,10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x,y)

ax.set_title("Sine Wave")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Figure and Axes
fig, ax = plt.subplots()

# Plot multiple lines on same Axes
ax.plot(x, y1, label="Sin")
ax.plot(x, y2, label="Cos")

# Labeling
ax.set_title("Sin and Cos Waves")
ax.set_xlabel("X Axis")
ax.set_ylabel("Values")

# Legend
ax.legend()

plt.show()
```

| Code Element               | Meaning                                    | Why It Matters                    |
| -------------------------- | ------------------------------------------ | --------------------------------- |
| `fig, ax = plt.subplots()` | Creates the canvas + plotting area         | Your control center               |
| `ax.plot()`                | Draws on THIS axes only                    | Prevents messy overlapping state  |
| `ax.set_*()`               | Title/labels assigned to the correct chart | Consistency in multi-plot figures |
| `ax.legend()`              | Uses the `label=` names automatically      | Clean legend without hacks        |


Subplots (Multiple Axes in One Figure)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

axes[0].plot(x, np.sin(x), color='blue')
axes[0].set_title("Sine Wave")

axes[1].plot(x, np.cos(x), color='red')
axes[1].set_title("Cosine Wave")

plt.tight_layout()
plt.show()
```

| Concept                         | Meaning                                | Why it matters                               |
| ------------------------------- | -------------------------------------- | -------------------------------------------- |
| `fig, axes = plt.subplots(...)` | axes becomes a **list-like container** | lets you control multiple plots individually |
| `axes[i]`                       | select the subplot                     | no accidental overwrites                     |
| `tight_layout()`                | fixes spacing                          | prevents cramped, ugly visuals               |


This is the foundation for dashboards, EDA visualization boards, and any professional report.

Customizing Every Part of a Plot

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(x, np.sin(x), color='purple', linewidth=3, linestyle='--', marker='o', markersize=6)

ax.set_title("Styled Sine Wave", pad=15)
ax.set_xlabel("X Axis Label")
ax.set_ylabel("Y Axis Label")

# Custom tick locations
ax.set_xticks([0, 2, 4, 6, 8, 10])
ax.set_yticks([-1, -0.5, 0, 0.5, 1])

# Annotate a point
ax.annotate("Peak", xy=(np.pi/2, 1), xytext=(2, 1.3),
            arrowprops=dict(arrowstyle="->", color="black"))

plt.show()
```
