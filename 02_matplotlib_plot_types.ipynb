# Matplotlib Plot Types (Practical Usage)

This notebook covers core plot types in Matplotlib, focusing on:
- When to use each plot
- How to interpret them
- Clean, minimal styling for professional visualization

We will cover:
1. Line Plot
2. Scatter Plot
3. Bar & Grouped Bar Plot
4. Histogram & Distribution View
5. Box Plot (Spread & Outliers)


1) Line Plot (Trends)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y, linewidth=2)

ax.set_title("Line Plot - Trend Over Continuous Range")
ax.set_xlabel("X")
ax.set_ylabel("sin(x)")

plt.show()
```

When to Use Line Plots

Time series

Smooth continuous functions

Anything with natural progression

Never use a line plot for categorical data.
Only continuous.

2) Scatter Plot (Relationships)

```python
np.random.seed(0)
x = np.random.randn(100)
y = x * 2 + np.random.randn(100) * 0.5

fig, ax = plt.subplots()
ax.scatter(x, y, alpha=0.7)

ax.set_title("Scatter Plot - Relationship Between Variables")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
```

Interpretation

Tight cluster  strong correlation

Weak spread  weak correlation

Pattern shape  possible non-linear trend

3) Bar Plot (Comparing Categories)

```python
categories = ["A", "B", "C", "D"]
values = [4, 7, 1, 8]

fig, ax = plt.subplots()
ax.bar(categories, values)

ax.set_title("Bar Plot - Category Comparison")
ax.set_xlabel("Category")
ax.set_ylabel("Value")

plt.show()
```

Rule

Use when comparing discrete groups.
Not continuous data.

4) Grouped Bar Plot (Very Useful in Business Data)

```python
labels = ["Q1", "Q2", "Q3", "Q4"]
sales_2023 = [120, 150, 130, 170]
sales_2024 = [140, 160, 135, 190]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
ax.bar(x - width/2, sales_2023, width, label="2023")
ax.bar(x + width/2, sales_2024, width, label="2024")

ax.set_title("Grouped Bar - Yearly Comparison")
ax.set_xlabel("Quarter")
ax.set_ylabel("Sales")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

This is used constantly in:

Revenue reports

Hiring numbers

KPI dashboards

5) Histogram (Distribution)

```python
data = np.random.randn(500)

fig, ax = plt.subplots()
ax.hist(data, bins=30, edgecolor="black")

ax.set_title("Histogram - Distribution Shape")
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")

plt.show()
```

Key

bins changes the story.

Too few  misleading.

Too many  noise.

6) Box Plot (Outliers & Spread)

```python
data = np.random.randn(100)

fig, ax = plt.subplots()
ax.boxplot(data)

ax.set_title("Box Plot - Spread & Outliers")
ax.set_ylabel("Value")

plt.show()
```

Interpretation Skill:

Box middle: median

Box size: IQR (spread)

Dots: outliers (important in ML preprocessing)
