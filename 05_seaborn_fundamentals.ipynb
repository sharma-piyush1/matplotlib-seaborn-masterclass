# Seaborn Fundamentals (Statistical Visualization Layer)

Seaborn builds on Matplotlib and is designed to:
- Reveal patterns
- Highlight relationships
- Show distributions clearly
- Work easily with Pandas DataFrames

We will use built-in datasets and compare plot types intentionally.


1) Load Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips.head()
```

2) Distribution Plots (Core Insight: Spread and Shape)
Histogram + KDE (Smooth distribution curve)

```python
sns.histplot(tips["total_bill"], kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

When to use:

Understand numeric variable shape (normal, skewed, bimodal)

Detect extreme values before modeling

3) Categorical â†’ Numeric (Comparing Groups)

Boxplot

```python
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Total Bill by Day")
plt.show()
```

Violinplot (same but shows density)

```python
sns.violinplot(data=tips, x="day", y="total_bill")
plt.title("Distribution Shape by Day")
plt.show()
```

Interpretation Skill (critical):

Box center = median

Box height = IQR (spread)

Whiskers = variability

Outliers = interesting anomalies, not â€œerrorsâ€ by default

4) Scatter + Regression line

Used constantly in ML feature relationship checking.

```python
sns.regplot(data=tips, x="total_bill", y="tip")
plt.title("Tip vs Total Bill")
plt.show()
```

Meaning:

upward line = positive relationship

slope steepness = strength

point spread = noise

5) Categorical Count Plot

```python
sns.countplot(data=tips, x="day")
plt.title("Count of Observations by Day")
plt.show()
```

### Interpretation Notes
- Use distribution plots to understand numeric data shape.
- Use box/violin to compare groups.
- Use scatter + regression to detect linear relationships.
- Use countplot to validate category balance.
