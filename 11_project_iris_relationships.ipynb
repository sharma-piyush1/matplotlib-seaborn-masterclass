# Project: Feature Relationship Analysis (Iris Dataset)

Objective:
Analyze how features relate to each other to identify:
- Which features are most informative for classification
- Which features are redundant
- Where separation between classes is strongest

Dataset: Seaborn "iris"


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.describe()
iris['species'].value_counts()
```

If class distribution is balanced → life is easy.
If not → prepare for headaches.
Here, it's balanced. Good.

```python
sns.pairplot(iris, hue="species", diag_kind="kde")
```

Interpretation You MUST Write:

Which pair of features visually separates classes best?

Which pairs overlap and offer no separation?

sns.heatmap(iris.corr(numeric_only=True), annot=True, cmap="coolwarm", center=0)


```python
sns.heatmap(iris.corr(numeric_only=True), annot=True, cmap="coolwarm", center=0)
```

Interpretation:

If two features are highly correlated → one is redundant in a model.

Redundant features cause:

Overfitting

Weight instability

Longer training time

No additional predictive power

```python
sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species")
plt.title("Petal Features - Best Natural Class Separation")
plt.show()
```

This plot will slam your face with the truth:

Petal length + petal width = clean separability

Sepal features = hot garbage for classification

This is why:

Every Iris classifier gets >95% accuracy without trying.

Because the data is already separable.

### Key Feature Insights
1. Petal length and petal width provide the strongest class separation.
2. Sepal width and sepal length show significant overlap and are weak predictors.
3. The dataset is linearly separable, which means simpler models (Logistic Regression, Linear SVM) perform well.
4. Including redundant features is unnecessary and may reduce model generalization clarity.

The most predictive feature pair in the Iris dataset is petal length and petal width, as these two variables cleanly separate species with minimal overlap. The weakest feature pair is sepal length and sepal width, which show significant overlap and therefore contribute little to class discrimination. The dataset is linearly separable, especially when using petal-based features. As a result, simple linear models such as Logistic Regression or a Linear SVM are appropriate and effective for classification. For model building, it is recommended to exclude redundant sepal features and focus on petal-based predictors to achieve high accuracy with minimal complexity.
