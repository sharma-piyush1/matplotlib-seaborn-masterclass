```python
import matplotlib.pyplot as plt
import seaborn as sns

# Apply seaborn global theme
sns.set_theme(
    style="whitegrid",      # adds grid
    context="notebook",      # balanced label sizes
)

# Apply consistent figure & font sizes
plt.rcParams.update({
    "figure.figsize": (10, 6),
    "axes.titlesize": 16,
    "axes.labelsize": 14,
    "xtick.labelsize": 12,
    "ytick.labelsize": 12,
    "lines.linewidth": 2,
})
```

### Why Style Configuration Matters
Consistent visuals are a signal of professionalism.
These settings ensure:
- clean layout
- readable labels
- no childish colors
- consistent figure sizing across the whole repo
