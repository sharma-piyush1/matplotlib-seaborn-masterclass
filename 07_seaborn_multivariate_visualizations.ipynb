# Multivariate Visualization (Finding Patterns Across Multiple Variables)

This notebook focuses on visualizations that:
- Compare multiple variables at once
- Reveal patterns, clusters, and correlations
- Understand interaction effects between features

You will use these plots heavily in EDA (Exploratory Data Analysis) and before modeling.


1) Pairplot (Global Relationship Overview)

This is your fastest multi-variable pattern scanner.
Useful before machine learning.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.pairplot(tips, hue="sex", diag_kind="kde")
plt.show()
```

What to look for:

Are relationships linear, curved, or random?

Are there separations between groups?

Which variables are redundant (correlated)?

2) Heatmap (Correlation Insight)

This is where machine learning feature selection begins.

```python
corr = tips.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

Interpretation Notes:

Values close to 1  strong positive relation

Values close to -1  strong negative relation

Values near 0  unrelated

Example: If total_bill and tip are strongly correlated, dont treat them as independent.

3) Jointplot (Relationship + Distribution in One View)

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="reg")
plt.show()
```

Why this matters:

Scatter shows relationship

Histograms on sides show distribution shapes

Regression line shows direction

This is basically three plots in one.

4) Catplot (When categories interact with other dimensions)

```python
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    kind="bar"
)
plt.show()
```

This answers real questions:

Do men spend more than women on weekends vs weekdays?

Is spending pattern stable?

This is the bread and butter of business analytics.

5) Swarm + Box Combined (High detail, still readable)

```python
sns.boxplot(data=tips, x="day", y="tip")
sns.swarmplot(data=tips, x="day", y="tip", color="black", size=3)
plt.show()
```

This gives:

Distribution shape (box)

Actual individual values (swarm)

Interpretation clarity

### Interpretation Framework
For every multivariate visualization, ask:
1. Which variables move together?
2. Which categories differ meaningfully?
3. Which relationships are strong enough to matter?
4. Where are the unexpected patterns?

This is what separates EDA from decoration.
